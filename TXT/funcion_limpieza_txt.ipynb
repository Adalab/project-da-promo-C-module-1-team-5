```python
def limpieza_txt (self): # Vamos a quitar los puntos y comas de la lista y convertir cada string en lista.
        
        lista_listas = []
        for i in self:
            
            lista_intermedia = i.split(";")
            
            lista_listas.append(lista_intermedia)
            
    #def espacios(self):# Vamos a quitar los espacios de mas en cada string.
        
        lista_final = []
        for sublista in lista_listas[1:]:

            lista_almancenado_strings = []

            for palabra in sublista:
                palabra = palabra.strip()
                lista_almancenado_strings.append(palabra)
                
            lista_final.append(lista_almancenado_strings)

    #def under(self):

        lista_finalcasi = []

        for sublista in lista_final[:]:
    
            lista_almancenado_under = []
            for palabra in sublista:
                i = palabra.replace('<','under')
                lista_almancenado_under.append(i)
        
            lista_finalcasi.append(lista_almancenado_under)

#def null(self):

        lista_finalTOTAL = []
    
        for sublista in lista_finalcasi[:]:

            lista_almancenado_null = []
            for palabra in sublista:
       
                x = palabra.replace('null','NULL')
                lista_almancenado_null.append(x)
    
            lista_finalTOTAL.append(lista_almancenado_null)

        lista_tupla = []

        for z in lista_finalTOTAL:
            lista_starwars = tuple(z) 
            lista_tupla.append(lista_starwars)   

        return lista_tupla
```
