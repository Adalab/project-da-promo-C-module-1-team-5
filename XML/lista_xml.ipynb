```python
import re
import os
import xml.etree.ElementTree as ET

tree = ET.parse('../data/data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
xml_list
```

```python
def cambiar_genero(lista_xml):

    for dict in (lista_xml):
        if dict.get('gender') == '0':
         dict['gender'] = 'Man'
        elif dict.get('gender') == '1':
         dict['gender'] = 'Woman'
        elif dict.get('gender') == '2':
         dict['gender'] = 'Nonbinary'
        elif dict.get('gender') == '3':
         dict['gender'] = 'Prefer not to say'
        elif dict.get('gender') == '4':
         dict['gender'] = 'Prefer to self-describe'
        else:    
         print(f"There is an error, the unlisted values is : {dict.get('gender')}")
    return "Género cambiado"   
        
```

```python
cambiar_genero(xml_list)
```
